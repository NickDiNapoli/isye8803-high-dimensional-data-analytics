## Mathematical notation of what to solve 

$min_b\lVert  y - Zb  \rVert _2^2 + \sum_{j=1}^{10} \lVert  b_j  \rVert_2$

where Z is matrix of b-splines coefficients and $b_j$ are the coefficients of for group j so this second term above is the group lasso.

```python
# from example code
def BSplineBasis(x: np.array, knots: np.array, degree: int) -> np.array:
    
    nKnots = knots.shape[0]
    lo = min(x[0], knots[0])
    hi = max(x[-1], knots[-1])
    
    augmented_knots = np.append(np.append([lo]*degree, knots), [hi]*degree)
    DOF = nKnots + degree + 1 # DOF = K+M, M = degree+1
    spline = BSpline(augmented_knots, np.eye(DOF), degree, extrapolate=False)
    B = spline(x)
    return B
```

```python
# training set

deg = 3
k = 8
x_train_reduced = []
for i in range(10):
    X = np.linspace(0, 1, x_train[0].shape[1])
    knots = np.linspace(0, 1, k)
    B = BSplineBasis(X, knots, deg)[:,:-2]
    #print(B.shape, x_train[i].shape)
    Bcoef = np.linalg.lstsq(B, x_train[i].T, rcond=None)[0].T
    #print(B.shape, Bcoef.shape)
    x_train_reduced.append(Bcoef)
    #y_splines = B@Bcoef.T

x_train_reduced = np.asarray(x_train_reduced)

X = np.linspace(0, 1, y_train.shape[1])
knots = np.linspace(0, 1, k)
B = BSplineBasis(X, knots, deg)[:,:-2]
print(B.shape, y_train.shape)
Bcoef = np.linalg.lstsq(B, y_train.T, rcond=None)[0].T
print(B.shape, Bcoef.shape)
y_train_reduced = Bcoef
y_splines = B@Bcoef.T
```

```python
# test set

deg = 3
k = 8
x_test_reduced = []
for i in range(10):
    X = np.linspace(0, 1, x_test[0].shape[1])
    knots = np.linspace(0, 1, k)
    B_test = BSplineBasis(X, knots, deg)[:,:-2]
    #print(B.shape, x_train[i].shape)
    Bcoef_test = np.linalg.lstsq(B_test, x_test[i].T, rcond=None)[0].T
    #print(B.shape, Bcoef.shape)
    x_test_reduced.append(Bcoef_test)
    #y_splines = B@Bcoef.T

x_test_reduced = np.asarray(x_test_reduced)

X = np.linspace(0, 1, y_test.shape[1])
knots = np.linspace(0, 1, k)
B_test = BSplineBasis(X, knots, deg)[:,:-2]
print(B_test.shape, y_test.shape)
Bcoef_test = np.linalg.lstsq(B_test, y_test.T, rcond=None)[0].T
print(B_test.shape, Bcoef_test.shape)
y_test_reduced = Bcoef_test
y_splines_test = B_test@Bcoef_test.T
```
