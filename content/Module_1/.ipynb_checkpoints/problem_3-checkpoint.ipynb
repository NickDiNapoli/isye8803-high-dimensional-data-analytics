```python
X_raw = data_df.index.values
X = np.linspace(0, 1, X_raw.shape[0])
Y = np.squeeze(data_df.to_numpy())
n = Y.shape[0]

plt.figure(figsize=(15,10))
plt.plot(X_raw, Y, '.', label='Observations')

MSE_cubic = []

for k in range(4, 14):
    
    knots = np.linspace(1/(k+1), k/(k+1), k)
    #print(X)

    H = []
    H.append(np.ones((X.shape[0], 1)))
    H.append(X.reshape(n, -1))
    H.append(X.reshape(n, -1)**2)
    H.append(X.reshape(n, -1)**3)

    #print(H[0].shape, H[1].shape, H[2].shape, H[3].shape)
    for ks in knots:
        H.append(np.maximum((X-ks)**3, 0).reshape(n, -1))

    H = np.hstack(H)
    #print(H.shape, Y.shape)
    B = np.linalg.lstsq(H, Y, rcond=None)[0]
    #print(B.shape)
    #B2 = np.linalg.solve(H.T@H, H.T@Y)
    #B3 = np.linalg.inv(H.T@H)@H.T@Y
    y_cubic = H@B
    plt.plot(X*n + 1950, H@B, label=f'{k} knot estimate')
    
    # LOOCV
    loo = LeaveOneOut()
    errs = []
    for trg, tst in loo.split(X):

        errs.append(y_cubic[tst]-Y[tst])
        
    MSE_cubic.append(sum([err**2 for err in errs]))
    
plt.legend()
plt.grid()
plt.title('Cubic splines estimate of coal electrical energy data')
plt.show()

plt.plot(MSE_cubic)
plt.xlabel("# knots")
plt.title("MSE")
```

## B-splines

```python
# from example code
def BSplineBasis(x: np.array, knots: np.array, degree: int) -> np.array:
    '''Return B-Spline basis. Python equivalent to bs in R or the spmak/spval combination in MATLAB.
    This function acts like the R command bs(x,knots=knots,degree=degree, intercept=False)
    Arguments:
        x: Points to evaluate spline on, sorted increasing
        knots: Spline knots, sorted increasing
        degree: Spline degree. 
    Returns:
        B: Array of shape (x.shape[0], len(knots)+degree+1). 
    Note that a spline has len(knots)+degree coefficients. However, because the intercept is missing 
    you will need to remove the last 2 columns. It's being kept this way to retain compatibility with
    both the matlab spmak way and how R's bs works.

    If K = length(knots) (includes boundary knots)
    Mapping this to R's bs: (Props to Nate Bartlett )
    bs(x,knots,degree,intercept=T)[,2:K+degree] is same as BSplineBasis(x,knots,degree)[:,:-2]
    BF = bs(x,knots,degree,intercept=F) drops the first column so BF[,1:K+degree] == BSplineBasis(x,knots,degree)[:,:-2]
    '''
    nKnots = knots.shape[0]
    lo = min(x[0], knots[0])
    hi = max(x[-1], knots[-1])
    
    augmented_knots = np.append(np.append([lo]*degree, knots), [hi]*degree)
    DOF = nKnots + degree + 1 # DOF = K+M, M = degree+1
    spline = BSpline(augmented_knots, np.eye(DOF), degree, extrapolate=False)
    B = spline(x)
    return B
```

```python
plt.figure(figsize=(15,10))
plt.plot(X_raw, Y, '.', label='Observations')
deg = 3
MSE_splines = []

for k in range(4, 14):

    knots = np.linspace(0, 1, k)
    B = BSplineBasis(X, knots, deg)[:,:-2]
    #print(B.shape)
    Bcoef = np.linalg.lstsq(B, Y.T, rcond=None)[0].T
    y_splines = B@Bcoef
    #print(y_splines.shape)
    plt.plot(X*n + 1950, B@Bcoef, label=f'{k} knot estimate')
    
    # LOOCV
    loo = LeaveOneOut()
    errs = []
    for trg, tst in loo.split(X):

        errs.append(y_splines[tst]-Y[tst])
        
    MSE_splines.append(sum([err**2 for err in errs]))
    

plt.legend()
plt.grid()
plt.title('B splines estimate of coal electrical energy data')
plt.show()

plt.plot(MSE_splines)
plt.xlabel("# knots")
plt.title("MSE")
```

## Kernel regression

```python
# Gaussian kernel

def kerf(z):
    return np.exp(-z*z/2)*(2*np.pi)**-0.5
```

```python
#X = np.arange(69)
bandwidths = np.arange(.001, 1.001, 0.001)
MSEs = []
for w in bandwidths:
    loo = LeaveOneOut()
    errs = []
    for trg, tst in loo.split(X):
        z = kerf((X[tst]-X[trg])/w)
        #print(z)
        #print(trg, tst.shape)
        yke = np.average(Y[trg], weights=z)
        errs.append(Y[tst]-yke)
        
    MSEs.append(sum([err**2 for err in errs]))
    
MSEs = np.array(MSEs).squeeze()
w_star = bandwidths[np.argmin(MSEs)]
print(w_star)

plt.plot(bandwidths, MSEs, label='Validation Curve')
plt.plot(w_star, min(MSEs),'r+', label='Lowest LOO MSE')
plt.title('Leave-One-Out Validation Curve, Kernel Estimator')
plt.xlabel('Kernel Bandwidth')
plt.ylabel('MSE')
plt.legend()
plt.show()

plt.plot(MSEs[0:10])
```

```python
w_star_test = .03
yy, yy_test = [], []
for xx in X:
    z = kerf((xx-X)/w_star)
    yy.append(np.average(Y, weights=z))
    
    z = kerf((xx-X)/w_star_test)
    yy_test.append(np.average(Y, weights=z))

plt.figure(figsize=(10,6))
plt.plot(X_raw, Y, '.', label='Observations')
plt.plot(X_raw, yy, 'r', label='RBF Kernel Regression, lambda=optimal')
plt.plot(X_raw, yy_test, 'k', label='RBF Kernel Regression, lambda=.03')
plt.legend()
plt.title('RBF Kernel Smoother')
plt.show()
```
