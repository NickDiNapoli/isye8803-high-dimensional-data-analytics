```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.interpolate import BSpline
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import plot_confusion_matrix
from scipy.ndimage import gaussian_filter
```

```python
with open('./ECG200TRAIN') as f:
    points = f.readlines()

train = []
for n in range(len(points)):
    
    train.append([float(p) for p in points[n].split(',')])

train = np.asarray(train)
x_train = train[:,1:]
y_train = train[:,0]
print(x_train.shape, y_train.shape)
    
plt.figure(figsize=(10,6))
for signal in x_train:
    plt.plot(signal)
    
plt.title("Training data")
```

```python
with open('./ECG200TEST') as f:
    points = f.readlines()

test = []
for n in range(len(points)):
    
    test.append([float(p) for p in points[n].split(',')])

test = np.asarray(test)
x_test = test[:,1:]
y_test = test[:,0]
print(x_test.shape, y_test.shape)
    
plt.figure(figsize=(10,6))
for signal in x_test:
    plt.plot(signal)
    
plt.title("Testing data")
```

## B-splines

```python
# from example code
def BSplineBasis(x: np.array, knots: np.array, degree: int) -> np.array:

    nKnots = knots.shape[0]
    lo = min(x[0], knots[0])
    hi = max(x[-1], knots[-1])
    
    augmented_knots = np.append(np.append([lo]*degree, knots), [hi]*degree)
    DOF = nKnots + degree + 1 # DOF = K+M, M = degree+1
    spline = BSpline(augmented_knots, np.eye(DOF), degree, extrapolate=False)
    B = spline(x)
    
    return B
```

```python
x = np.linspace(0, 1, x_train.shape[1])
#print(X)
knots = np.linspace(0, x_train.shape[0], 8)
B = BSplineBasis(x, knots, degree=3)[:,:-2]

print(B.shape, x_train.T.shape)
Bcoef = np.linalg.lstsq(B, x_train.T, rcond=None)[0].T
Bcoef_test = np.linalg.lstsq(B, x_test.T, rcond=None)[0].T
print(Bcoef.shape, Bcoef_test.shape)
# classification
rf = RandomForestClassifier()
rf.fit(Bcoef, y_train)
print(Bcoef.shape)

pred = rf.predict(Bcoef_test)
conf = confusion_matrix(y_test, pred)

conf = pd.DataFrame(conf, index=['Class 0', 'Class 1'], columns=['Predicted 0', 'Predicted 1'])
print('Confusion Matrix - Splines\n', conf)

plot_confusion_matrix(rf, Bcoef_test, y_test)  
```

## FPCA

```python
X = np.vstack([x_train, x_test])
mu_hat = gaussian_filter(X.mean(0),3)

diffs = X-mu_hat
Cov = np.cov(diffs.T)
l, psi = np.linalg.eigh(Cov)
print(psi.shape)
PCs = psi[:, -2:]
print(PCs.shape)
FPC_scores = diffs.dot(PCs)

rf = RandomForestClassifier()
rf.fit(FPC_scores[0:100, :], y_train)
pred = rf.predict(FPC_scores[100:200])
conf = confusion_matrix(y_test, pred)
conf = pd.DataFrame(conf, index=['Class 0', 'Class 1'], columns=[
                    'Predicted 0', 'Predicted 1'])
print('Confusion Matrix - FPCA\n', conf)

plot_confusion_matrix(rf, FPC_scores[100:200], y_test) 
```
