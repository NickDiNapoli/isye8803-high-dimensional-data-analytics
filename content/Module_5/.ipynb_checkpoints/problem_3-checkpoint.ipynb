```python
import numpy as np
import matplotlib.pyplot as plt
from random import sample
```

```python
M_0 = np.random.rand(50, 50) 
mask = M_0 < 0.5
M_0[mask] = 0
m = np.count_nonzero(M_0)

print(M_0)
print(m)
```

```python
# adapted from example code 

# %Initialization
Y = np.zeros((50, 50))
delta = 50*50/m
tau = 250

vec = []
err = []
for i in range(500):
    u, s, vh = np.linalg.svd(Y)
    s_t = np.maximum(s-tau, 0)
    Z = (u[:, :50]*s_t)@vh
    P = M_0-Z
    P[mask] = 0
    Y0 = Y.copy()
    Y = Y0 + delta*P

reconstruction_error = np.linalg.norm(Y - M_0, ord='fro') / np.linalg.norm(M_0, ord='fro')
```
