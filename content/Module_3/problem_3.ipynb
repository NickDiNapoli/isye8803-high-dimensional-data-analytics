```python
import matplotlib.pyplot as plt
import tensorly as tl
from itertools import product
import numpy as np
import tensorly.decomposition
from scipy import io
import h5py
```

```python
data = io.loadmat('./heatT.mat')
print(type(data), data.keys())
#print(data)

T1 = data['T1'][0][0][0]
T2 = data['T2'][0][0][0]
T3 = data['T3'][0][0][0]
print(T1.shape)

plt.figure(figsize=(20,8))
# example heat transfer
for t in range(10):
    plt.subplot(3, 10, t+1)
    plt.imshow(T1[:,:,t], cmap='inferno')
    
    plt.subplot(3, 10, 10+t+1)
    plt.imshow(T2[:,:,t], cmap='inferno')
    
    plt.subplot(3, 10, 20+t+1)
    plt.imshow(T3[:,:,t], cmap='inferno')
    
plt.tight_layout()
```

```python
# helper functions from example code

def enforceBdy(S):
    ''' Enforces the boundary conditions on S, the temperature values on the domain's grid points'''
    S[:, 0] = 1
    S[:, -1] = 1
    S[0, :] = 1
    S[-1, :] = 1
    return S

def Laplace(T):
    '''Computes the Laplacian operator, del-squared on the temperature data'''
    tmp_x, tmp_y = np.gradient(T, dx, dy)
    tmp_x, _ = np.gradient(tmp_x, dx)
    _, tmp_y = np.gradient(tmp_y, dy)
    return tmp_x+tmp_y
```

```python
X = T2 # toggle 
nx, ny, nt = X.shape

#S = np.zeros((ny, nx))
#S = enforceBdy(S)
# print(S)

# CP decomposition modified from example code
err = []
for i in range(1,11):
    CP_Heat = tl.decomposition.parafac(X, i)
    reconstructed = tl.kruskal_to_tensor(CP_Heat)  
    err.append(((X-reconstructed)**2).sum())
    
AIC = [2*e + 2*(i+1) for i, e in enumerate(err)]

idxmin = np.argmin(AIC)
R = idxmin+1
min_AIC = AIC[idxmin]

print(f"Rank: {R}", f"AIC: {min_AIC}")
plt.plot(np.arange(1,11), AIC)
plt.plot(R, min_AIC, 'r*', markersize=10)
plt.xlabel('R')
plt.ylabel('AIC')
plt.show()

P = tl.decomposition.parafac(X, R)
P_T3 = tl.decomposition.parafac(T3, R)
print(P)
P_recon = tl.cp_to_tensor(P)
P_T3_recon = tl.cp_to_tensor(P_T3)
print("Tensor difference: ", np.linalg.norm(P_recon - P_T3_recon))

plt.figure()
plt.plot(P[1][2])
plt.xlabel('Time')
plt.legend(['Component ' +str(i) for i in range(1,R+1)], fontsize=5)
plt.title(f'CP decomposition')
plt.show()

plt.figure(figsize=(20,8))
for i in range(idxmin+1):
    A = P[1][0][:,i]
    B = P[1][1][:,i]
    XY = np.outer(A,B)
    plt.subplot(1, 10, i+1)
    plt.imshow(XY, cmap='inferno')
    plt.title(f'Spatial Comp {i}x{i}')
plt.tight_layout()
plt.show()
```

```python
# Tucker Decomposition modified from example code
AIC = {}
for i,j,k in product(range(1,10),repeat=3):
    tucker_heat = tl.decomposition.tucker(X,[i,j,k])
    reconstructed = tl.tucker_to_tensor(tucker_heat)  
    e = ((X-reconstructed)**2).sum()
    n_params = i*j*k+np.prod(tucker_heat[1][0].shape)+np.prod(tucker_heat[1][1].shape)+np.prod(tucker_heat[1][2].shape)
    AIC[(i,j,k)] = 2*e + 2*n_params

#print(list(AIC.values()))
plt.figure()
plt.plot(list(AIC.values()))

minIJK = min(AIC, key=AIC.get)
# minIJK = (3,3,3)
minAIC = AIC[minIJK]
print(minIJK, minAIC)
tucker = tl.decomposition.tucker(X, minIJK)
tucker_T3 = tl.decomposition.tucker(T3, minIJK)
I,J,K = minIJK

tucker_recon = tl.tucker_to_tensor(tucker)
tucker_T3_recon = tl.tucker_to_tensor(tucker_T3)
print("Tensor difference: ", np.linalg.norm(tucker_recon - tucker_T3_recon))

plt.figure()
plt.plot(tucker[1][-1])
plt.xlabel('Time/frame')
plt.legend(['Component ' +str(i) for i in range(1,K+1)])
plt.show()
#print(i)
plt.figure(figsize=(20,8))

count = 1
for i,j in product(range(I),range(J)):

    A = tucker[1][0][:,i]
    B = tucker[1][1][:,j]
    XY = np.outer(A,B)
    plt.subplot(1, 10, count)
    plt.imshow(XY,cmap='inferno')
    plt.title(f'Spatial Comp {i+1}x{j+1}')
    count += 1
plt.tight_layout()
plt.show()
```
