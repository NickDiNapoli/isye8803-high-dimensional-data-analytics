```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import skimage
from skimage import io, transform, color
import scipy.ndimage as ndi
from sklearn.cluster import k_means
from skimage.segmentation import watershed
from skimage.feature import peak_local_max
```

```python
img = io.imread('./facemask-3.jpg')
print(img.dtype, img.shape)
plt.figure(figsize=(10,5))
plt.imshow(img)
plt.title("Original image")
plt.savefig("orignal_image.png")

m, n, _ = img.shape
img = img[:, 0:int(n/2), :]
plt.figure(figsize=(10,5))
plt.imshow(img)
plt.title("Cropped image")

img_gray = color.rgb2gray(img)*255
print(img_gray.shape)
plt.figure(figsize=(10,5))
plt.imshow(img_gray, cmap='gray')
plt.title("Cropped image grayscale")
plt.savefig("cropped_image.png")

img_norm = img_gray/img_gray.max()
plt.figure(figsize=(10,5))
plt.imshow(img_norm, cmap='gray')
plt.title("Cropped image normalized")
```

```python
plt.figure(figsize=(12,6))
plt.hist(img_gray.ravel(), bins = 256, color='black')
plt.title("Grayscale histogram")
plt.savefig("grayscale_histogram.png")
```

```python
facemask = img_gray[215:365, 250:500]
print(facemask.shape)
plt.figure(figsize=(10,5))
plt.imshow(facemask, cmap='gray')
plt.title("Thresholded cropped facemask image")

img_bw = np.where(facemask > 0.2*255, 1., 0.)
plt.figure(figsize=(10,5))
plt.imshow(img_bw, cmap='gray')
plt.title("Thresholded cropped facemask image, p=0.2")
plt.savefig("small_crop.png")

img_linear = 255 - img_bw
plt.figure(figsize=(10,5))
plt.imshow(img_linear, cmap='gray')
```

```python
x = np.where(img_bw == 0)[0] + 215
y = np.where(img_bw == 0)[1] + 250

mask = np.ones(shape=img_gray.shape)
for i,j in zip(x,y):
    mask[i,j] = 0
    
mask *= 255
plt.figure(figsize=(10,5))
plt.imshow(mask, cmap='gray')
plt.title("Extracted mask")
plt.savefig("extracted_mask.png")
```

```python
# Canny algorithm

guass_filter = (1/1115) * np.array([[1, 4, 7, 10, 7, 4, 1], 
                                    [4, 12, 26, 33, 26, 12, 4], 
                                    [7, 26, 55, 71, 55, 26, 7], 
                                    [10, 33, 71, 91, 71, 33, 10], 
                                    [7, 26, 55, 71, 55, 26, 7], 
                                    [4, 12, 26, 33, 26, 12, 4],
                                    [1, 4, 7, 10, 7, 4, 1]])

#img_smooth = ndi.convolve(img, guass_filter)
img_smooth = cv2.filter2D(src=img_gray, ddepth=-1, kernel=guass_filter)

term1 = np.diff(img_smooth, axis=0)
term2 = np.roll(np.diff(img_smooth, axis=0), -1, axis=1)
#print(term1)
#print(term2)
dS_dx = 0.5 * (term1 + term2)
#dS_dx = dS_dx[:,:-1,:]
dS_dx = dS_dx[:,:-1]

term1y = np.diff(img_smooth, axis=1)
term2y = np.roll(np.diff(img_smooth, axis=1), -1, axis=0)
#print(term1)
#print(term2)
dS_dy = 0.5 * (term1y + term2y)
#dS_dy = dS_dy[:-1,:,:]
dS_dy = dS_dy[:-1,:]

G = np.sqrt(np.square(dS_dx) + np.square(dS_dy))
theta = np.arctan(dS_dy / dS_dx)
theta = np.nan_to_num(theta, nan=0)

theta_discrete = np.piecewise(theta,
            [((theta > (-1/8)*np.pi) & (theta <= (1/8)*np.pi)), 
             ((theta > (1/8)*np.pi) & (theta <= (3/8)*np.pi)), 
             ((theta > (-3/8)*np.pi) & (theta <= (-1/8)*np.pi)), 
             ((((theta > (3/8)*np.pi) & (theta <= (1/2)*np.pi))) | (((theta >= (-1/2)*np.pi) & (theta <= (-3/8)*np.pi))))], 
            [0, (1/4)*np.pi, (-1/4)*np.pi, (1/2)*np.pi])

m, n = G.shape
phi = G
for i in range(1, m-1):
    for j in range(1, n-1):
        
        # check all conditions
        if theta_discrete[i,j] == 0:
            p1 = (i, j-1)
            p2 = (i, j+1)
            if G[i,j] < G[p1] or G[i,j] < G[p2]:
                phi[i,j] = 0
                
        if theta_discrete[i,j] == (1/4)*np.pi:
            p1 = (i+1, j-1)
            p2 = (i-1, j+1)
            if G[i,j] < G[p1] or G[i,j] < G[p2]:
                phi[i,j] = 0
                
        if theta_discrete[i,j] == (-1/4)*np.pi:
            p1 = (i-1, j-1)
            p2 = (i+1, j+1)
            if G[i,j] < G[p1] or G[i,j] < G[p2]:
                phi[i,j] = 0
                
        if theta_discrete[i,j] == (1/2)*np.pi:
            p1 = (i-1, j)
            p2 = (i+1, j)
            if G[i,j] < G[p1] or G[i,j] < G[p2]:
                phi[i,j] = 0
                
m, n = phi.shape
#print(m,n)
E = np.zeros(shape=phi.shape)
t1, t2 = 5, 15
count = 1
while count != 0:
    count = 0
    for i in range(1, m-1):
        for j in range(1, n-1):
            #print(phi[i,j])
            if phi[i,j] > t2 and E[i,j] == 0:
                E[i,j] = 1
                count += 1
                
            if phi[i,j] > t1 and E[i,j] == 0:
                for neighbor in [E[i-1,j-1], E[i-1,j], E[i-1,j+1], E[i,j-1], E[i,j], E[i,j+1], E[i+1,j-1], E[i+1,j], E[i+1,j+1]]:
                    if neighbor == 1:
                        E[i,j] = 1
                        count += 1


plt.figure(figsize=(10,5))
plt.imshow(phi, cmap='gray') 
#print(phi)                       
plt.figure(figsize=(10,5))
plt.imshow(E, cmap='gray')
```
